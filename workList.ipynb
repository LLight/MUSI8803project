We have three file types, with different data processing steps needed for each type. In the final corpus, all files will be converted into the YCAC format (CSV with one row per chord, and all chords in normal form representation).

 
The following columns indicate whether each file processing step is complete:
1) phraseAnalysis: annotation of phrase boundaries in CSV file 
 - applies to all file types
 
2) fileConversion: file has been converted from pdf or mxl into music21
 - applies to musicXML (parsed using music21) and pdf (scanned using photoscore)
 
3) getChords: vertical slicing and identification of normal form chords in music21 
 - applies to musicXML and pdf
