```python
xmlFiles=["Copland_Piano_Blues_No_3.mxl","Messiaen_Regard_du_Pere.mxl","Part_Fuer_Alina.mxl","Poulenc_Les_chemins_de_lamour.mxl",
          "Rautavaara_Etude_op42_no1_Thirds.mxl","Schnittke_Tango.mxl","Sculthorpe_Sea_Chant.mxl","Takemitsu_Romance.mxl",
         "Allard_5diversions.xml","Bachlund_Commedia_Dellarte.xml","Beatty_Musing_for_Piano.xml","Simpson_Invention_no6.xml"]
len(xmlFiles)
```

```python
#parse a single file to test the process
filename=xmlFiles[1] 
c = converter.Converter()
c.parseFile(filename)
s = c.stream
s.show()
```

```python
#parse all mxl files downloaded from MuseScore
for piece in (xmlFiles):
    print(piece)
    c = converter.Converter()
    c.parseFile(piece)
    s = c.stream
```
